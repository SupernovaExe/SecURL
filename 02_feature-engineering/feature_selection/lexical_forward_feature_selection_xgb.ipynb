```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
dataset = pd.read_csv('final_unbalanced_with_lexical.csv')
dataset.head()
```

```python
features = dataset.iloc[:, 1:]
url_type = dataset.iloc[:, 0]
```

```python
features.head()
```

```python
url_type.head()
```

```python
features_train, features_test, url_type_train, url_type_test = train_test_split(features, url_type, test_size = 0.3)
```

```python
forward_feature_selection = SequentialFeatureSelector(XGBClassifier(), k_features = 76, forward = True,
                                                      floating = False, verbose = 2, scoring = 'accuracy', cv = 5).fit(features_train, url_type_train)
```
