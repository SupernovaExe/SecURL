```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import feature_generation_lexical_function
from sklearn.preprocessing import LabelEncoder

dataset = pd.read_csv("final_unbalanced_noFeatures.csv")

dataset["url_type"] = dataset["type"].replace({
    'benign':0,
    'defacement':1,
    'phishing':2,
    'malware':3,
});
```

```python
dataset['url_scheme'] = dataset['url'].apply(lambda x: feature_generation_lexical_function.url_scheme(x))
dataset['get_tld'] = dataset['url'].apply(lambda x: feature_generation_lexical_function.get_tld(x))
```

```python
get_tld = dataset['get_tld'].to_list()
url_scheme = dataset['url_scheme'].to_list()
```

```python
le = LabelEncoder()
dataset['get_tld'] = le.fit_transform(dataset['get_tld'])
dataset['url_scheme'] = le.fit_transform(dataset['url_scheme'])

scheme_cat = dataset['url_scheme'].to_list()
tld_cat = dataset['get_tld'].to_list()

scheme_lookup = {}
tld_lookup = {}
i = 0
for scheme in url_scheme:
    scheme_lookup[scheme] = scheme_cat[i]
    i += 1
i = 0
for tld in get_tld:
    tld_lookup [tld] = tld_cat[i]
    i += 1
```

```python
scheme_lookup_pd = pd.DataFrame(list(scheme_lookup.items()), columns=['Key', 'Value'])
scheme_lookup_pd.to_csv("scheme_lookup.csv", encoding='utf-8', index=False)
```

```python
tld_lookup_pd = pd.DataFrame(list(tld_lookup.items()), columns=['Key', 'Value'])
tld_lookup_pd.to_csv("tld_lookup.csv", encoding='utf-8', index=False)
```
